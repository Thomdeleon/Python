```python
def fibTail(current, last, iterator):
    if iterator == 1:
        return current
    return fibTail(current + last, current, iterator-1)
```

```python
%timeit fibTail(1, 0, 5)
```

```python
def fibb(current, last, iterator):
    if iterator > 1:
        return fibb(current + last, current, iterator-1)
    return current
```

```python
%timeit fibb(1, 0, 5)
```
